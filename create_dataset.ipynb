```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('pataaka.csv')
```

```python
df
```

```python
df["region_shape_attributes"] = df["region_shape_attributes"].apply(lambda x : dict(eval(x)) )
df2 = df["region_shape_attributes"].apply(pd.Series )
```

```python
df2
```

```python
df2['x_min'], df2['y_min']=df2['x'], df2['y']
df2['x_max'], df2['y_max']=df2['x']+df2['width'], df2['y']+df2['height']
```

```python
df2['width'], df2['height']=128, 128
```

```python
result = pd.concat([df, df2], axis=1).drop(['region_shape_attributes', 'name', 'x', 'y', 'file_attributes', 'region_count', 'region_id', 'region_attributes'], axis=1)
result['class']='Pataaka'
```

```python
result
```

```python
result.to_csv("pataaka.csv",index=False, columns=result.columns)
```
